```python
import nibabel as nib
import matplotlib.pyplot as plt
from ipywidgets import interact



def view_slice(nii_path,z):
    img = nib.load(nii_path)
    data = img.get_fdata()
    plt.imshow(data[:, :, z], cmap='gray')
    plt.title(f"Slice {z}")
    plt.axis('off')
    plt.show()

nii_path = '/home/hasselnot/ML/SAM-Med3D/test_data/amos_val_toy_data/imagesVa/amos_0013.nii.gz'
interact(view_slice, nii_path=nii_path, z=(0, data.shape[2]-1))
```
